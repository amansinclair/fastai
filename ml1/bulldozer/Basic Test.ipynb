```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_train = pd.read_pickle("./data/Train.pkl")
```

```python
y = df_train["SalePrice"].copy()
```

```python
df_train.drop(["SalePrice"], axis=1, inplace=True)
```

```python
def convert_to_cat(df):
    for col in df.columns:
        if df[col].dtype == "object":
            df[col] = df[col].astype("category")
            df[col] = df[col].cat.codes
```

```python
convert_to_cat(df_train)
```

```python
def fill_numerical(df):
    for col in df.columns:
        if df[col].dtype == "float64":
            df[col].fillna(df[col].median(), inplace=True)
```

```python
fill_numerical(df_train)
```

```python
df_train["saledate"] = df_train["saledate"].apply(lambda x : int(x.timestamp()))
```

```python
train_size = int(len(df_train) * 0.8)
```

```python
X_train, y_train = df_train.iloc[:train_size], y[:train_size]
X_test, y_test = df_train.iloc[train_size:], y[train_size:]
```

```python
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.metrics import mean_squared_log_error as msle
from sklearn.metrics import r2_score
```

```python
rfr = RFR(n_estimators=40, oob_score=True, max_features=0.5, bootstrap=True, max_samples=0.2)
rfr.fit(X_train, y_train)
preds = rfr.predict(X_test)
print("msle", msle(y_test, preds), "r2", r2_score(y_test, preds), "oob", rfr.oob_score_)
```

```python
idxs = np.flip(np.argsort(rfr.feature_importances_))
```

```python
plt.figure(figsize=(10,10))
sns.barplot(rfr.feature_importances_[idxs], X_train.columns[idxs])
plt.show()
```
