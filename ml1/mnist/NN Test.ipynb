```python
import numpy as np
import matplotlib.pyplot as plt
import os
from ml1 import fit, Linear, Sequential, ReLU, LogSoftmax, NLLLoss
```

```python
folder = "data"
X_train = np.load(os.path.join(folder, "X_train.npy"))
y_train = np.load(os.path.join(folder,"y_train.npy"))
X_valid = np.load(os.path.join(folder,"X_valid.npy"))
y_valid = np.load(os.path.join(folder,"y_valid.npy"))
```

```python
my_net = Sequential(Linear(28*28, 10), ReLU(), LogSoftmax())
my_loss = NLLLoss()
```

```python
fit(my_net, X_train, y_train, X_valid, y_valid, my_loss, batch_size=20, n_epochs=20)
```

```python
def view_some_preds(net, X, y):
    idxs = np.random.choice(X.shape[0], size=16, replace=False)
    output = net(X[idxs])
    preds = np.argmax(output, axis=1)
    fig, ax = plt.subplots(4,4, figsize=(15, 10))
    plot_n = 0
    for i in range(4):
        for j in range(4):
            idx = idxs[plot_n]
            ax[i, j].set_title(str(int(preds[plot_n])) + "(" + str(int(y[idx])) + ")")
            ax[i, j].imshow(X[idx].reshape(28, 28), cmap="gray")
            ax[i, j].axis("off")
            plot_n += 1
    plt.show()
```

```python
view_some_preds(my_net, X_train, y_train)
```
